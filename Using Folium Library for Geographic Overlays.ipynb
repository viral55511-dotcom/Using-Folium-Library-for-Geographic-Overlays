```python
#Let us read the dataset
import pandas as pd

df = pd.read_csv("Indicators", on_bad_lines='skip', engine='python')

print(df.shape)
df.head()
```

```python
# Select CO2 emissions for all countries in 2011

hist_indicator = r'CO2 emissions \(metric'
hist_year = 2011

mask1 = df['IndicatorName'].str.contains(hist_indicator, na=False)
mask2 = df['Year'] == hist_year

co2_2011 = df[mask1 & mask2]

print(co2_2011.shape)
co2_2011.head()
```

```python
stage = df[
    (df['IndicatorName'] == 'CO2 emissions (metric tons per capita)') &
    (df['Year'] == 2011)
]

print(stage.shape)
stage.head()
```

```python
#Setup the data for plotting
#Create a data frame with just the country codes and the values we want to be
#plotted.
plot_data = stage[['CountryCode','Value']]
plot_data.head()
```

```python
# Label for the legend
hist_indicator = stage.iloc[0]['IndicatorName']
print(hist_indicator)
```

```python
#Visualize CO2 emissions per capita using Folium
#Folium provides interactive maps with the ability to create sophisticated
# overlays for data visualization
#Setup a folium map at a high-level zoom.
map = folium.Map(location=[100, 0], zoom_start=1.5)
```

```python
import folium

country_geo = 'world-countries.json'  # adjust path

m = folium.Map(location=[20, 0], zoom_start=2)

folium.Choropleth(
    geo_data=country_geo,
    data=plot_data,
    columns=['CountryCode', 'Value'],
    key_on='feature.id',
    fill_color='YlGnBu'
).add_to(m)

m.save("co2_map.html")
```
